```python
import glob
from pathlib import Path
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import soundfile as sf
import librosa
import librosa.display
import IPython.display as display
```

```python
def read_ogg_file(path, file):
    """ Read ogg audio file and return numpay array and samplerate"""
    data, samplerate = sf.read(path+file)
    return data, samplerate
```

```python
INPUT_DIR = '../input/train_short_audio/logshr/*'

paths = list(glob.glob(INPUT_DIR))
```

# データ読み込み

## data load

```python
path = paths[0]

data, sr = sf.read(path)
print(f'data shape: {data.shape}')
print(f'sampling rate: {sr}')
print(f'num frame: {len(data) / sr}')
```

## 音声信号

```python
plt.plot(data)
plt.title('waveform')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()

display.Audio(path)
```

# 特徴量抽出
## STFT
### STFT変換

```python
n_fft = 512 # 窓の長さ
hop_length = n_fft//4 # 窓の移動幅
nyquist_hz = sr / 2 # ナイキスト周波数
bin_hz = 1 + n_fft // 2 # 周波数ビンの数（周波数分解能）
bin_time = sr / hop_length # 1秒あたり時刻ビンの数 (時間分解能)
width_bin_hz = nyquist_hz / bin_hz # 周波数ビンの幅
width_bin_time = 1000 / bin_time # 時刻ビンの幅


print(f'nyquist Hz: {nyquist_hz}')
print(f'number of bin Hz: {bin_hz}')
print(f'number of bin time: {bin_time}')
print(f'width of bin Hz: {width_bin_hz}')
print(f'width of bin time: {width_bin_time}')

stft_data = librosa.stft(data, n_fft=n_fft, hop_length=hop_length)
```

```python
print(stft_data.shape)
# 複素数の2次元配列
print(stft_data)
```

### 極形式に変換

```python
# 極形式に変換し、強度を使う
# どちらも同じ処理をしてる
S, phase = librosa.magphase(stft_data)
SS = np.abs(stft_data)
print(np.all(S==SS))
```

### 強度のヒストグラム

```python
# 行が周波数、列が時間を表し、要素は強度
print(S.shape)
sdf = pd.DataFrame(S.flatten())

plt.hist(sdf[0].values, bins=20, log=True)
plt.title('power histogram')
plt.show()
```

### 位相の散布図

```python
print(phase.shape)

pdf = pd.DataFrame(phase.flatten()).sample(n=1000)

# 位相の散布図
plt.figure(figsize=(3, 3))
plt.scatter(np.real(pdf), np.imag(pdf))
plt.show()
```

### 強度をdB単位へ変換

```python
Sdb = librosa.amplitude_to_db(S)
print(Sdb.shape)
sdf = pd.DataFrame(Sdb.flatten())

plt.hist(sdf[0].values, bins=20, log=True)
plt.title('power histogram')
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2, figsize=(8, 8))
librosa.display.specshow(S, sr=sr, x_axis='time', y_axis='hz', ax=axes[0][0])
librosa.display.specshow(Sdb, sr=sr, x_axis='time', y_axis='hz', ax=axes[0][1])

librosa.display.specshow(S, sr=sr, x_axis='time', y_axis='log', ax=axes[1][0])
librosa.display.specshow(Sdb, sr=sr, x_axis='time', y_axis='log', ax=axes[1][1])

plt.tight_layout()
plt.show()
```

### 実装してみる
#### 波形を適当な長さに分割

```python
data, sr = sf.read(path)

t = np.arange(0, len(data) / sr, 1/sr)

# 音声波形の中心部分（定常部）を切り出す
center = len(data) // 2

n_fft = 512 # 窓の長さ
hop_length = n_fft//4 # 窓の移動幅
nyquist_hz = sr / 2 # ナイキスト周波数
bin_hz = 1 + n_fft // 2 # 周波数ビンの数（周波数分解能）
bin_time = sr / hop_length # 1秒あたり時刻ビンの数 (時間分解能)
width_bin_hz = nyquist_hz / bin_hz # 周波数ビンの幅
width_bin_time = 1000 / bin_time # 時刻ビンの幅

print(f'nyquist Hz: {nyquist_hz}')
print(f'number of bin Hz: {bin_hz}')
print(f'number of bin time: {bin_time}')
print(f'width of bin Hz: {width_bin_hz}')
print(f'width of bin time: {width_bin_time}')

cut_data = data[int(center - n_fft//2): int(center + n_fft//2)]
time = t[int(center - n_fft//2): int(center + n_fft//2)]

print(cut_data.shape, time.shape)
plt.plot(time, cut_data)
# plt.ylim(-0.0025, 0.001)
```

#### 窓関数をかける

```python
# ハミング窓をかける
hamming = np.hamming(len(cut_data))
cut_data *= hamming

# 両端のシグナルが弱まった
plt.plot(time, cut_data)
# plt.ylim(-0.0025, 0.001)
```

#### FFT

```python

spectrum = np.abs(np.fft.fft(cut_data, n_fft))[:n_fft//2]
fscale = np.fft.fftfreq(n_fft, d=1.0 / sr)[:n_fft//2]
print(spectrum.shape)
print(fscale.shape)
plt.plot(spectrum)
```

## MFCC
### メル尺度

```python
f0 = 700
m0 = 2595

f = np.arange(0, 32000) # 周波数の配列
m = m0 * np.log1p(f / f0) # 周波数fをメル尺度mに変換

plt.plot(f, m)
plt.xlabel('Hz')
plt.ylabel('mel scale')
plt.show()
```
